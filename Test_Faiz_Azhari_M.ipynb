```python
x = (data.drop(['Unnamed: 0','Country', 'Gender', 'Game', 'Total Follower', 'Broadcast Hours', 'PaidStarPerWatchedHour', 'PaidStarPerWatchedHour.1', 'MBTI_grouped'], axis=1)).values
```

```python
y = np.nan_to_num(data['PaidStarPerWatchedHour'].values)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
prediksi = LinearRegression()
prediksi.fit(x_train, y_train)
```

```python
y_prediksi = prediksi.predict(x_test)
print(y_prediksi)
```

```python
prediksi.predict([x[1]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_prediksi)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.scatter(y_test, y_prediksi)
plt.xlabel('Aktual')
plt.ylabel('Prediksi')
```

```python
tabel_prediksi = pd.DataFrame({'Aktual':y_test, 'Prediksi':y_prediksi, 'Beda':y_test-y_prediksi})
```
